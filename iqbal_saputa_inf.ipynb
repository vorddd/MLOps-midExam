# Libraries

```python
# Library Dataframe
import pandas as pd

# Library Numerical Data
import numpy as np

# Library Model Saving
import joblib

# Library Random Data
from scipy.stats import randint
```

# Data Inference

## **Load Model**

```python
# Memuat model pipeline terbaik
from pathlib import Path
MODEL_PATH = Path('models') / 'best_model_pipeline.joblib'
with open(MODEL_PATH, 'rb') as file_1:
  model_pipeline = joblib.load(file_1)
```

## **Make Random Data**

```python
# Number of rows
num_rows = 10

# Generate random data within realistic ranges
dataFrame = {
    'Customer_care_calls': np.random.randint(2, 8, num_rows),
    'Cost_of_the_Product': np.random.randint(96, 311, num_rows),
    'Prior_purchases': np.random.randint(2, 6, num_rows),
    'Discount_offered': np.random.randint(1, 20, num_rows),
    'Weight_in_gms': np.random.randint(1492, 5018, num_rows),
    'Product_importance': np.random.choice(['low', 'medium', 'high'], num_rows)
}

random_data = pd.DataFrame(dataFrame)
random_data
```

## **Model Prediction**

## **Interpretasi Hasil**
Tabel berikut menunjukkan data sintetis yang diuji beserta prediksi model (Tepat Waktu / Tidak Tepat Waktu).

```python
# Model Prediction
a = model_pipeline.predict(random_data)
prediction_label = np.where(a == 1, 'Tepat Waktu', 'Tidak Tepat Waktu')

result_df = random_data.copy()
result_df['Prediction'] = prediction_label
result_df
```
